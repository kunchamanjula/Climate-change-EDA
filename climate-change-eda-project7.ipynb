# 1.Show the temperature changes happened  in 2020 over different countries?

```python
plt.figure(figsize=(10,5))
sns.histplot(climate["F2020"],color="green",kde=True)  # using seaborn to plot histogram
plt.title("Climate changes in 2020 across different countries",fontsize=14 )
plt.xlabel("temperature indicator in celsius")
plt.ylabel("count of countries")
plt.show()
```

# creating new dataframe 

```python
year_list = []
for i in range(1963, 2023):
    year = f"F{i}"
    year_list.append(year)
    
melt = pd.melt(climate, id_vars = ["Country"], value_vars = year_list)
melt.rename(columns={"variable":"Year", "value":"Tempurate_Change"}, inplace = True)
melt['Year'] = melt['Year'].str.replace(r'F', '', regex=True)
melt['Year'] = melt['Year'].astype(int)
```

```python
melt
```

#  3.plot the graph to show the countries which has highest temperature changes

```python
fig_bar = px.bar(melt.sort_values("Tempurate_Change", ascending = False).head(10), x='Country',
                 y='Tempurate_Change', color='Tempurate_Change', color_continuous_scale = 'orrd',
                 title='Countries With The Highest Temperature Increase (Top 10)', text="Country",
                 labels={'Tempurate_Change': 'Tempurate Change (°C)'})

fig_bar.update_layout(
    xaxis=dict(showgrid=False),
    yaxis=dict(showgrid=False),
    plot_bgcolor='rgba(0,0,0,0)'
)
fig_bar.show()
```

# 4.plot the graph based on the climatic changes on happen in these decades ?

```python
# categorzing the years into decades for better visuvalization
# Creating a new DataFrame with year as a column and a new decade column
years = [col for col in climate.columns if col.startswith('F')]
global_avg_temp_change = climate[years].mean()

# Convert years from string 'FYYYY' to integer YYYY for plotting
int_years = [int(year[1:]) for year in years]

decades= pd.melt(climate, id_vars=['Country'], value_vars=years, var_name='Year', value_name='TempatureChange') 
decades['Decade'] = decades['Year'].apply(lambda x: x[1:4] + "0s")  # Group years into decades
plt.figure(figsize=(12, 8))
sns.boxenplot(x='Decade', y='TempatureChange', data=decades)
plt.title('Distribution of Temperature Changes by Decades', fontsize=16)
plt.xlabel('Decade', fontsize=14)
plt.ylabel('Temperature Change (°C)', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

# 5.plot the graph to show the mean teampature change for every year in dataset?

```python
mean_tempchange = melt.groupby("Year").agg({"Tempurate_Change":"mean"}).reset_index()
```

```python
fig = px.line(mean_tempchange.sort_values("Year", ascending=False), x='Year', y='Tempurate_Change',
              title='Mean Tempurate Change for each Year', markers=True,
              labels={'Tempurate_Change': 'Tempurate Change (°C)'}, color_discrete_sequence=['#B62644'])

fig.update_layout(
    xaxis=dict(showgrid=False),
    yaxis=dict(showgrid=False),
    plot_bgcolor='rgba(0,0,0,0)',
)
fig.show()
```

# 6.provide the Max and Min climatic  Change By Year?

```python
melt['Tempurate_Change'] = pd.to_numeric(melt['Tempurate_Change'], errors='coerce')
max_tempurate = melt.loc[melt['Tempurate_Change'].idxmax()]
min_tempurate = melt.loc[melt['Tempurate_Change'].idxmin()]
print("-"*50)
print("Max Tempurate: ",max_tempurate)
print("-"*50)
print("Min Tempurate: ",min_tempurate)
```

# 7.plot the graph to show The country with the maximum temperature change for each year?

```python
top_countries_year = melt.groupby(["Year", "Country"])["Tempurate_Change"].mean().reset_index()
top_countries_year = top_countries_year.groupby("Year").apply(lambda x: x.nlargest(1, "Tempurate_Change")).reset_index(drop=True)
top_countries_year = top_countries_year[top_countries_year["Year"] > 1999]
```

```python
fig = px.bar(top_countries_year.sort_values("Year", ascending=False), x='Year', y='Tempurate_Change', color='Tempurate_Change',
              color_continuous_scale = 'orrd', title='The country with the maximum temperature change for each year', text="Country",
              labels={'Tempurate_Change': 'Tempuarete Cahnge (°C)'})

fig.update_layout(
    xaxis=dict(showgrid=False),
    yaxis=dict(showgrid=False),
    plot_bgcolor='rgba(0,0,0,0)', 
)
fig.show()
```

# 8.plot heatmap to visualize temperature changes across different countries over a selected range of years?

```python
selected_years = ['F2012', 'F2014', 'F2016', 'F2018', 'F2019', 'F2020', 'F2021']
heatmap_data = climate.set_index('Country')[selected_years]

# Creating a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(heatmap_data, cmap='coolwarm')
plt.title('Heatmap of Temperature Changes (2010, 2012, 2014, 2016, 2018, 2020, 2022)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Country', fontsize=14)
plt.show()
```
